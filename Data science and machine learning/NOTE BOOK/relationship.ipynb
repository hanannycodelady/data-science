```python
df = pd.read_excel(r'C:\Users\STUDENT\Desktop\WITU COHORT 3\Year one\Semester one\Data science and machine learning/DATA/Dataset.xlsx.xlsx')
print(df) 
```

```python
import matplotlib.pyplot as plt
```

```python
Weight=df["Weight"]
```

```python
Height=df["Height"]
```

```python
plt.plot( Weight , Height,"o" )
```

```python
plt.plot( Weight , Height,"o",alpha=0.02 )
```

```python
plt.plot( Weight , Height,"o" )
plt.title("A scatter diagram showing height and weight of cohort 3")
plt.xlabel("Weight (kgs)")
plt.ylabel("Height(cm)")
```

```python
import numpy as np
```

```python
# How to put the line of the best feat
slope, intercept =np.polyfit(Height,Weight, 1)
plt.plot( Weight , Height,"o" )
plt.plot(Height, slope * Height + intercept)
plt.title("A scatter diagram showing height and weight of cohort 3")
plt.xlabel("Weight (kgs)")
plt.ylabel("Height(cm)")
```

# violin plot

```python
import seaborn as sns
```

```python
df = pd.read_excel(r'C:\Users\STUDENT\Desktop\WITU COHORT 3\Year one\Semester one\Data science and machine learning/DATA/Dataset.xlsx.xlsx')
print(df) 
```

```python
weight=df["Weight"]
```

```python
age=df["Age"]
```
